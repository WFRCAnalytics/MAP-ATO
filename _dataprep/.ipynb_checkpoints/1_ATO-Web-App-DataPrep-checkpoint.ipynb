### Use Arcpy to convert dbfs to csvs

```python
from arcpy import TableToTable_conversion
from arcpy import env

# Create function that converts dbfs to csvs using arcpy
def dbfToCSV(dbf, csv):
    TableToTable_conversion(dbf, os.path.dirname(csv), os.path.basename(csv))

def getCSVFilename(year):
    return os.path.join(dirIntermediate, strATORootName + str(year) + '.csv')
    
# Allow arcpy to overwrite output files
env.overwriteOutput = True

# store location of data folder
results_folder = os.path.join(os.getcwd(), 'Results')

# convert dbfs to csvs
for year in listYear:
    display("Convert " + str(year) + " (dbf to csv)")
    dbf_year = os.path.join(dirATOdata, strATORootName + str(year) + '.dbf')
    csv_year = getCSVFilename(year)
    dbfToCSV(dbf_year, csv_year)
```

```python
def unique(list1):
    # intilize a null list
    unique_list = []

    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)

    return unique_list
```

```python
#calculate averages

df_Avg = pd.DataFrame()

sdf_TAZ = pd.DataFrame.spatial.from_featureclass(filenameTAZshp)

for y, year in enumerate(unique(listAvgYear)):
    #display(year)
    
    df_ATO = pd.read_csv(getCSVFilename(year))
    
    #rename composite fields
    df_ATO = df_ATO.rename(columns={"COMP_TRAN":"TRAN_COMP", "COMP_AUTO":"AUTO_COMP"})
    
    #display(df_ATO)
    
    for mode in listMode:
        for category in listCategory:
            #display(str(year) + '_' + mode + "_" + category)
            
            df_ATOf = df_ATO[['CO_TAZID', mode + "_" + category]]
            df_ATOf.columns = ("Z","V")
            
            df_ATOf.columns = ("CO_TAZID","VALUE")
                        
            df_ATOf['LABEL'] = round(df_ATOf['VALUE'] / 1000,0).astype(int).astype('str') + "k"
            
            #display(df_ATOf)
            
            df_TAZa_ATOf = pd.DataFrame.merge(df_TAZa,df_ATOf,on="CO_TAZID")
                        
            df_SAa = df_TAZa_ATOf.groupby(['SAID'], as_index=False).agg({"CO_TAZID":[np.size],"VALUE":[np.mean]})
            df_SAa.columns = ('SAID','TAZCNT','AVG')
            
            df_SAa['AVG'] = df_SAa['AVG'].astype('int64')
            
            #display(df_SAa)
            #display(df_ATOf_SAa.describe())
            
            for sa in listSA: 
                
                saAvg = df_SAa[df_SAa['SAID']==sa]['AVG'].values[0]
                
                df_Avg = df_Avg.append({'Name'           : 'YEAR_' + str(year) + '_' + mode + '_' + category + '_' + sa,
                                        'Average'        :  saAvg
                                        },ignore_index=True)


  
```

```python
for y, year in enumerate(listYear):
    display(year)
    
    df_ATO = pd.read_csv(getCSVFilename(year))
    
    #rename composite fields
    df_ATO = df_ATO.rename(columns={"COMP_TRAN":"TRAN_COMP", "COMP_AUTO":"AUTO_COMP"})
    
    #display(df_ATO)
    
    for mode in listMode:
        for category in listCategory:
            #display(str(year) + '_' + mode + "_" + category)
            
            strName = 'YEAR_' + str(year) + '_AVG_' + str(listAvgYear[y]) + '_' + mode + '_' + category
            
            df_ATOf = df_ATO
            
            df_ATOf = df_ATO[['CO_TAZID', mode + "_" + category]]
            df_ATOf.columns = ("Z","V")
            
            #export df to json to use in chart prep
            df_ATOf.to_json(os.path.join(dirResults, strName + ".json"),orient='records',double_precision=0)

            df_ATOf.columns = ("CO_TAZID","VALUE")
                        
            df_ATOf['LABEL'] = round(df_ATOf['VALUE'] / 1000,0).astype(int).astype('str') + "k"
                       
            for sa in listSA: 
                
                #strName = 'YEAR_' + str(year) + '_' + mode + '_' + category + '_' + sa + '.json'
                #display(strName)
                
                #get average from average year                
                _avgname = 'YEAR_' + str(listAvgYear[y]) + '_' + mode + '_' + category + '_' + sa
                display(_avgname)
                
                saAvg = df_Avg[df_Avg['Name']==_avgname].Average.sum() #use sum to get value, but should be only one record
                                        
                display(saAvg)
                
                #df_ATOf['PCT_' + sa] = df_ATOf['VALUE'] / saAvg
                #df_ATOf['BIN_' + sa] = pd.cut(df_ATOf['PCT_' + sa], listBins, labels=False)                
                
                df_ATOf['AVG_' + sa] = saAvg
                df_ATOf['BIN_' + sa] = pd.cut(df_ATOf['VALUE'] / saAvg, listBins, labels=False)
                
                #df_Export = df_ATOf[['CO_TAZID','BIN','LABEL']].copy()
                #df_Export.columns = ('Z','B','L')
                #display(df_Export)
                
                #df_Export.to_json(os.path.join(dirResults, strName),orient='records',double_precision=2)
                
            #display(df_ATOf)
            
            sdf_TAZATO = pd.DataFrame.merge(sdf_TAZ,df_ATOf,on="CO_TAZID")
            filename = (gdb_ATO_Path + '\\' + strName)

            deleteIfExists(filename)
            display(filename)
            sdf_TAZATO.spatial.to_featureclass(filename)
            
```

```python
for gain in listGain:
    display(gain)
    
    year_B = gain[0]
    year_F = gain[1]
    
    df_ATO_B = pd.read_csv(getCSVFilename(year_B))
    df_ATO_F = pd.read_csv(getCSVFilename(year_F))

    #rename composite fields
    df_ATO_B = df_ATO_B.rename(columns={'COMP_TRAN':'TRAN_COMP', 'COMP_AUTO':'AUTO_COMP'})
    df_ATO_F = df_ATO_F.rename(columns={'COMP_TRAN':'TRAN_COMP', 'COMP_AUTO':'AUTO_COMP'})
    
    #display(year_B)
    #display(df_ATO_B)
    #display(year_F)
    #display(df_ATO_F)
    
    for mode in listMode:
        for category in listCategory:
            
            strName = 'GAIN_' + str(year_B) + 'to' + str(year_F) + '_' + mode + "_" + category
            
            df_ATOf_B = df_ATO_B[['CO_TAZID', mode + "_" + category]]
            df_ATOf_F = df_ATO_F[['CO_TAZID', mode + "_" + category]]

            df_ATOf_B.columns = ('CO_TAZID','VALUE_B')
            df_ATOf_F.columns = ('CO_TAZID','VALUE_F')
            
            df_ATOf_G = pd.DataFrame.merge(df_ATOf_B,df_ATOf_F,on='CO_TAZID')
            df_ATOf_G['VALUE'   ] =  df_ATOf_G['VALUE_F'] - df_ATOf_G['VALUE_B']
            #df_ATOf_G['VALUEPER'] = (df_ATOf_G['VALUE_F'] - df_ATOf_G['VALUE_B']) / df_ATOf_G['VALUE_B'] * 100
            
            df_ATOf_G['LABEL']    = round(df_ATOf_G['VALUE']   / 1000,0).astype(int).astype('str') + "k"
            #df_ATOf_G['LABELPER'] = round(df_ATOf_G['VALUEPER']      ,0).astype(int).astype('str') + "%"
            
            #compute gains
            df_TAZa_ATOf_G = pd.DataFrame.merge(df_TAZa,df_ATOf_G,on='CO_TAZID')
            
            df_SAa = df_TAZa_ATOf_G.groupby(['SAID'], as_index=False).agg({"CO_TAZID":[np.size],"VALUE":[np.mean]})
            df_SAa.columns = ('SAID','TAZCNT','AVG')
            
            df_SAa['AVG'] = df_SAa['AVG'].astype('int64')
            
            #display(df_ATOf_SAa)
            #display(df_ATOf_SAa.describe())
            
            for sa in listSA: 
                
                saAvg    = df_SAa[df_SAa['SAID']==sa]['AVG'   ].values[0]
                #saAvgPer = df_SAa[df_SAa['SAID']==sa]['AVGPER'].values[0]
                
                df_Avg = df_Avg.append({'Name'           : 'GAIN_' + str(year_B) + 'to' + str(year_F) + '_' + mode + '_' + category + '_' + sa,
                                        'Average'        :  saAvg
                                        },ignore_index=True)
                #display(saAvg)
            
                df_ATOf_G['BIN_'    + sa] = pd.cut(df_ATOf_G['VALUE'   ] / saAvg   , listBins, labels=False)
                #df_ATOf_G['BINPER_' + sa] = pd.cut(df_ATOf_G['VALUEPER'] / saAvgPer, listBins, labels=False)
                
                #df_Export = df_ATOf_G[['CO_TAZID','BIN','LABEL']]
                #df_Export.columns = ('Z','B','L')
                #display(df_Export)
                
                #df_Export.to_json(os.path.join(dirResults, strName),orient='records',double_precision=2)
                        
            sdf_TAZATO = pd.DataFrame.merge(sdf_TAZ,df_ATOf_G,on="CO_TAZID")
            filename = (gdb_ATO_Path + '\\' + strName)
            deleteIfExists(filename)
            sdf_TAZATO.spatial.to_featureclass(filename)
            
```
